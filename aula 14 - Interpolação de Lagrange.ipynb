```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
xa = np.array([0.1, 0.6])
ya = np.array([1.221, 3.320])

x = sp.symbols('x')
x0 = xa[0]
x1 = xa[1]
y0 = ya[0]
y1 = ya[1]

L1 = y0 * ((x - x1)/(x0 - x1)) + y1 * ((x - x0) / (x1 - x0))
print(L1)
x = 0.2
L1 = y0 * ((x - x1)/(x0 - x1)) + y1 * ((x - x0) / (x1 - x0))
print(L1)
```

Determine $L2(x)$

```python
xa = np.array([0.1, 0.6, 1.1])
ya = np.array([1.221, 3.320, 4.310])

x = np.linspace(0, 5, 100)
x0 = xa[0]
x1 = xa[1]
x2 = xa[2]
y0 = ya[0]
y1 = ya[1]
y2 = ya[2]

L2 = y0 * ((x - x1)/(x0 - x1) * (x - x2)/(x0 - x2)) + y1 * ((x - x0) / (x1 - x0) * (x - x2)/(x1 - x2)) + y2 * ((x - x2) / (x1 - x2) * (x - x1)/(x2 - x1))
#print(L2)
```

```python
def f(x): return y0 * ((x - x1)/(x0 - x1) * (x - x2)/(x0 - x2)) + y1 * ((x - x0) / (x1 - x0) * (x - x2)/(x1 - x2)) + y2 * ((x - x2) / (x1 - x2) * (x - x1)/(x2 - x1))
```

```python
def grafico(f):
    x = np.linspace(0, 5, 100)
    plt.plot(x, f(x))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico de Lagrange')
    plt.grid(True)
    plt.show()

grafico(f)
```
