### Raízes de Equações
Dada uma função f(x), $\bar {x}$ tal que f($\bar {x}$) = 0 é dito raíz da equação f(x) = 0

- Isolar as raízes

Determinar um intervalo que contenha uma raíz da equação:  $\bar {x} \in [a, b]$
- Refinar f($\bar {x}$)

Chegar ao valor com uma certa precisão.

#### Gráfico da Função

- Isolando graficamente

1) y = 2x³ - cos(x + 1) - 4

#### Importando as bibliotecas

Gráfico simples

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-100, 100, 300)
#print(x)

def f(x):
    return 2 * x ** 3 - np.cos(np.radians(x + 1)) - 4

#print(f(x))
plt.plot(x, f(x))
```

Gráfico detalhado

```python
x = np.linspace(-2.5, 2.5, 300)
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de f')
plt.grid(True)
plt.legend("x")
plt.show()
```

Gráfico a partir de função

```python
def grafico(f, a, b):
    x = np.linspace(a, b, 300)
    y = f
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico de f')
    plt.grid(True)
    plt.legend("x")
    plt.show()

grafico(f(x), -5, 5)
```
