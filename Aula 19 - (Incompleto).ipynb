```python
def PolinomioNewton(x, y):
    for e in x:
        if e < 0:
            x = x * (-1)
    
    x0, x1, x2 = x
    y0, y1, y2 = y

    fx01 = (y1 - y0) / (x1 - x0)
    fx12 = (y2 - y1) / (x2 - x1)
    fx012 = (fx12 - fx01) / (x2 - x0)
    
    resx = fx01 + fx012 * x1 + fx012 * x0
    res = fx01 * x0 - fx012 * (x0 * x1) 


    return f'{fx012}x² + {resx}x + {res}'
```
