> ## Integração Numérica

Queremos determinar $ \int_{a}^{b} f(x) \,dx $ por meio de uma técnica numérica. Uma motivação para esse estudo é, por exemplo, uma função para a qual não conhecemos uma primitiva , por exemplo, $ \frac{2} {\sqrt{\pi}} \int_{0}^{x} e^⁻t² \,dt $

Regra do Trapézio

$ \int_{a}^{b} f(x) \,dx $ = Área do trapézio

Se tomarmos $N(x)$ como uma aproximação de $f(x)$ no intervalo $[a, b]$:

$ \int_{a}^{b} f(x) \,dx \approx \int_{a}^{b} N(x) \,dx $

$ \int_{a}^{b} N(x) \,dx = \frac{f(a) + f(b)} {2} h $

Logo:

$ \int_{a}^{b} f(x) \,dx \approx \frac{f(a) + f(b)} {2} h $

h = n.&triangle; h

$> n < erro$

$ \int_{a}^{b} f(x) = \sum_{i=1}^{n} \int_{i} f(x) \,dx $

1.3) Funções para Integração Simples e Integração Composta

```python
def IntegracaoSimples(f, a, b):
    return (f(a) + f(b)) / 2 * (b - a)
```

```python
def IntegracaoComposta(f, a, b, n):
    resultado = 0
    h = (b - a) / n
    max = b - a
    a, b, j = 0, 0, 0
    while j < max:
        b += h
        resultado += (f(a) + f(b)) / 2 * h
        j += h
        a += h
    return resultado
```
