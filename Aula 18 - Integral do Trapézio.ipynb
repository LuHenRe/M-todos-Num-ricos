```python
import numpy as np
```

```python
def f(x): return np.sqrt(x - 1)
```

```python
def trapezio_s (fa, fb, h):
    I = h * (fa + fb) / 2 
    return I
```

```python
a = 1
b = 5
n = 4
h = (b - a) / n
print(h)
print(trapezio_s(a, b, h))
```

```python
x_barra = np.zeros(n + 1)

i = 0
while i < n + 1:
    x_barra[i] = a + i * h
    i += 1
```

```python
i = 0
while i < n + 1:
    print(x_barra[i], f(x_barra[i]))
    i += 1
```

```python
i = 0
I = 0

while i < n:
    I += trapezio_s(f(x_barra[i]), f(x_barra[i+1]), h)
    i += 1
    
print('Integral de f(x) de', a, 'a', b, ':', I)
```

### Função do Trapézio Composto

```python
def trapezio_composto(f, a, b, n):
    '''
    Calcula a integral f no intervalo [a, b] pela regra do trapézio.
    n = 1, trapézio simples
    '''
    def trapezio_s (fa, fb, h):
        I = h * (fa + fb) / 2 
        return I
    h = (b - a) / n
    x_barra = np.zeros(n + 1)

    i = 0
    while i < n + 1:
        x_barra[i] = a + i * h
        i += 1

    i = 0
    I = 0
    while i < n:
        I += trapezio_s(f(x_barra[i]), f(x_barra[i+1]), h)
        i += 1
    return I
```

```python
print(trapezio_composto(f, 1, 5, 4))
```
