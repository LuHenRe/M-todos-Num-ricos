#### Falsa Posição

```python
import numpy as np
import matplotlib.pyplot as plt
import Luiz_Henrique as lh
```

```python
def falsa_posicao(f, a, b, tol=1e-6, max_iter=100):
#Encontra a raiz, no intervalo [x0, x1], da equação definida em fParada: no máximo, max_iter, iterações ou diferença entre
#os limites do intervalo menor que tol.
    if f(a) * f(b) >= 0:
        print("ops! f(a) e f(b) devem ter sinais opostos")
        return None
    for i in range(max_iter):
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        if abs(f(c)) < tol:
            return c
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return None # não converge
```

```python
def funcao(x): return 2 * x ** 4 + 4 * x ** 3 - 11 * x ** 2 - 14 * x + 10
```

```python
raiz = falsa_posicao(funcao, -3, -2, tol=1e-6, max_iter=11)
print('raiz:', raiz)
print('função da raiz:', funcao(raiz))
```
