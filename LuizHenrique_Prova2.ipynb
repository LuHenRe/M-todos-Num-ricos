#### 1A) 
Obs: Foram feitos três gráficos para maior certeza.

```python
def f_lagrange(p): return 2 * p ** 2 + 3 * p - 2
def f_newton(p): return (3 * p ** 2 + 5 * p - 10) / 2
```

```python
a, b = 0, 1
x = np.linspace(a, b, 100)
plt.plot(x, f_lagrange(x), label='Método de Lagrange')
plt.plot(x, f_newton(x), label='Método de Newton')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.grid(True)
plt.legend()
plt.show()
```

```python
a, b = 0, 10
x = np.linspace(a, b, 100)
plt.plot(x, f_lagrange(x), label='Método de Lagrange')
plt.plot(x, f_newton(x), label='Método de Newton')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.grid(True)
plt.legend()
plt.show()
```

```python
a, b = 0, 100
x = np.linspace(a, b, 100)
plt.plot(x, f_lagrange(x), label='Método de Lagrange')
plt.plot(x, f_newton(x), label='Método de Newton')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.grid(True)
plt.legend()
plt.show()
```

A diferença de complexidade entre os métodos é observável desde pequenos pontos. O método de Newton é melhor desde um único ponto. Já em 10 pontos, é nítido a diferença de operações que se mantêm cada vez maior a cada ponto adicional. Portanto, o Método de Newton possui menos operações em todos os casos, e consequentemente, é mais rápido e vantajoso do que o Método de Lagrange.
