#### 2) $f(x) = x^3 -  5x² + 8x - 4$

```python
def funcao(x): return x ** 3 - 5 * x ** 2 + 8 * x - 4

def grafico(f, a, b):
    x = np.linspace(a, b, 100)
    plt.plot(x, f(x))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico da função f(x) = x³ - 5x² + 8x - 4')
    plt.grid(True)
    plt.show()
grafico(funcao, 0.75, 2.25)
```
